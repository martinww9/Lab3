```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
data_path = '/content/drive/MyDrive/20230714_Titulados_Ed_Superior_2022_WEB.csv'
data_path_2015 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2015_WEB.csv'
data_path_2016 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2016_WEB.csv'
data_path_2017 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2017_WEB.csv'
data_path_2018 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2018_WEB.csv'
data_path_2019 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2019_WEB.csv'
data_path_2020 = '/content/drive/MyDrive/20240404 Notas y Egresados EnseñanzaMedia 2020_WEB.csv'
data_path_2020_ren = '/content/drive/MyDrive/20210223_Rendimiento_2020_20210131_WEB.csv'
notas_2015 = pd.read_csv(data_path_2015, sep=';', encoding='latin-1')
notas_2016 = pd.read_csv(data_path_2016, sep=';', encoding='latin-1')
notas_2017 = pd.read_csv(data_path_2017, sep=';', encoding='latin-1')
notas_2018 = pd.read_csv(data_path_2018, sep=';', encoding='latin-1')
notas_2019 = pd.read_csv(data_path_2019, sep=';', encoding='latin-1')
notas_2020 = pd.read_csv(data_path_2020, sep=';', encoding='latin-1')
titulados_2020 = pd.read_csv(data_path, sep=';', encoding='latin-1')
rendimiento_2020 = pd.read_csv(data_path_2020_ren, sep=';', encoding='latin-1')
```

```python
#Pasamos a mayusculas para mantener consistencia con los demas datasets y poder hacer merge
titulados_2020.columns = titulados_2020.columns.str.upper()
titulados_2020.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Unir datasets por MRUN
merged_dataset2015 = pd.merge(notas_2015, titulados_2020, on='MRUN')
merged_dataset2016 = pd.merge(notas_2016, titulados_2020, on='MRUN')
merged_dataset2017 = pd.merge(notas_2017, titulados_2020, on='MRUN')
merged_dataset2018 = pd.merge(notas_2018, titulados_2020, on='MRUN')
merged_dataset2019 = pd.merge(notas_2019, titulados_2020, on='MRUN')
merged_dataset2020 = pd.merge(notas_2020, titulados_2020, on='MRUN')
merged_dataset2020_ren = pd.merge(rendimiento_2020, titulados_2020, on='MRUN')
```

```python
# Concatenar datasets por filas
notas_titulados = pd.concat([merged_dataset2015, merged_dataset2016,
                           merged_dataset2017, merged_dataset2018,
                           merged_dataset2019, merged_dataset2020, merged_dataset2020_ren], ignore_index=True)
notas_titulados
```

```python
notas_titulados.isnull().sum()
```

```python
notas_titulados['NIVEL_CARRERA_2'].value_counts()
```

```python
#Notamos que las notas estan en el formato '6,5' en vez de '6.5'
notas_titulados['PROM_NOTAS_ALU'] = notas_titulados['PROM_NOTAS_ALU'].str.replace(',', '.')

# Seleccionar características
features = [
    'GEN_ALU', 'RANGO_EDAD',
    'RBD', 'COD_REG_RBD', 'COD_PRO_RBD', 'COD_COM_RBD', 'COD_DEPROV_RBD',
    'PROM_NOTAS_ALU', 'SEM_ING_CARR_ORI', 'SEM_ING_CARR_ACT',
    'TIPO_INST_1', 'TIPO_INST_2', 'TIPO_INST_3', 'REGION_SEDE'
]

target = 'NIVEL_CARRERA_2'

# Filtrar dataset
X = notas_titulados[features]
y = notas_titulados[target]

```

```python
non_numeric_cols = X.select_dtypes(exclude=['number']).columns
X_encoded = pd.get_dummies(X, columns=non_numeric_cols)
```

```python
# Preprocesamiento
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Entrenar el modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluar el modelo
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
